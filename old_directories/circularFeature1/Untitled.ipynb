```python
import numpy as np
from scipy.special import jv as besseli
from matplotlib import pyplot as plt
```

```python
tuningSharpness=[0.1,1,2,5]

dRad = 0.01
tcenter = np.arange(dRad,2*np.pi+dRad,dRad)

for ts in tuningSharpness:
    vonmises = lambda x,u: np.exp(ts * np.cos(x-u))/(2*np.pi*besseli(0,ts))
    xx = vonmises(np.pi,tcenter)
    plt.plot(tcenter,xx)
    print(besseli(0,ts))
    #print(np.sum(xx*dRad))
```
